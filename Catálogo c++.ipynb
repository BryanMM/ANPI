# Método de la bisección

## Fórmula Matemática:
Sea $f$ una función real, definida y continua sobre un intervalo cerrado $[a,b]\in \mathbb{R}$\
Si se cumple que:
$$
f(a)\cdot f(b) \leq 0 \
$$
\
Entonces existe $x_{k} \in [a,b]$ tal que $f(x_{k})=0$, donde: \
\
$$
x_{k}=\frac{a+b}{2}
$$

### Valores Iniciales:

La función continua $f$, así como el intervalo $[a,b]$ que cumple $f(a) \cdot f(b) \leq 0$
